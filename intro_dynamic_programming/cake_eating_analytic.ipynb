# Analytical Solution to the Cake-Eating Problem

by [Manuel V. Montesinos](https://sites.google.com/view/manuelmontesinos), July 2020

Applying the all-in-one solution is very convenient for the simple cake-eating problem, as an analytical solution does exist. Yet if there were no analytical solution, the first-order conditions together with the agent's constraints would constitute an infinite number of equations that would have to be solved simultaneously. One way around this problem is to use dynamic programming, i.e., to decompose the all-in-one problem into many small sub-problems.

For that purpose, we can rewrite the agent's constraint as

$$a_0 = \sum_{v=0}^{t-1} c_v +c_t + \sum_{v=t+1}^\infty \Leftrightarrow a_0 - \sum_{v=0}^{t-1} c_v = c_t + \sum_{v=t+1}^{\infty} c_v.$$

We can interpret the left-hand side of the equation as how much out of the total value of the resource the agent has already consumed in the periods prior to time $t$:

$$a_0 = \sum_{v=0}^{t-1} c_v +c_t + \sum_{v=t+1}^\infty \Leftrightarrow a_0 - \sum_{v=0}^{t-1} c_v = a_t = c_t + \sum_{v=t+1}^{\infty} c_v.$$

This just tells us that the total consumption in $t$ as well as in any future period needs to be equal to the remaining resources $a_t$. We can see that $\sum_{v=t+1}^{\infty} c_v$ must hold, and therefore we can write the **dynamic budget constraint** or **transition equation** of the agent's resource as

$$a_t = c_t + a_{t+1} \Leftrightarrow a_{t+1} = a_t - c_t \ \forall \ t = 0,1,...,\infty.$$

This constraint tells us that what remains at $t+1$ is the amount of remaining resources $a_t$ at $t$ minus what the agent extracts $c_t$.

Let's now consider the utility function. We can define the sub-utility functions

$$U_t := \sum_{v=t}^{\infty} \beta^{v-t} u(c_v) = u(c_t) + \beta \sum_{v=t+1}^{\infty} \beta^{v-(t+1)} u(c_v) = u(c_t) + \beta U_{t+1}.$$

Let's take a look at the sub-optimization problem

$$\max_{\lbrace c_v \rbrace_{v=t}^{\infty}} U_t = \sum_{v=t}^{\infty} \beta^{v-t} u(c_v) \text{ s.t. } a_t = \sum_{v=t}^{\infty} \beta^{v-t} c_v.$$

We can set the Lagrangian for this problem and derive first-order conditions:

$$u'(c_t) = \beta^{v-t} u'(c_v) \Leftrightarrow c_v = (u')^{-1} \big[ \beta^{t-v} u'(c_t) \big] \forall v = t, t+1, ..., \infty,$$

where it is assumed that the marginal utility functions is invertible. Using the constraint, we get

$$a_t = \sum_{v=0}^{\infty} (u')^{-1} \big[ \beta^{-v} u'(c_t) \big].$$

Since $u'(c_t)$ is constant over all the elements of the infinite sum, the solution for $c_t$ is independent of $t$, but only depends on $a_t$. Hence,

$$c_t = f(a_t)$$ and $$c_v = (u')^{-1} \big[ \beta^{t-1} u'[f(a_t)] \big].$$

Plugging all this back into the utility function, 

$$\max_{\lbrace c_v \rbrace_{v=t}^{\infty}} U_t = \sum_{v=t}^{\infty} \beta^{v-t} u(c_v) \text{ s.t. } a_t = \sum_{v=t}^{\infty} \beta^{v-t} c_v$$

$$= \sum_{v=t}^{\infty} \beta^{v-t} u \Big\lbrace (u')^{-1} \big[ \beta^{t-v} u'[f(a_t)] \big] \Big\rbrace$$

$$\sum_{v=0}^{\infty} \beta^{v} u \Big\lbrace (u')^{-1} \big[ \beta^{-v} u'[f(a_t)] \big] \Big\rbrace =: V(a_t).$$

This results tells us that the solution of the sub-problem is independent of $t$, but only depends on $a_t$. In consequence, the maximum amount of utility the agent can derive from $t$ onwards under the assumption that only $a_t$ is left is independent of $t$ and can be subsumed in $V(a_t)$. This function $V$ is called the **value function**.

It is possible to show that

$$V(a_t) = \max_{\lbrace c_v \rbrace_{v=t}^{\infty}} U_t = u(c_t) + \beta U_{t+1} \text{ s.t. } \sum_{v=t}^{\infty} c_v = a_t$$

can be split into 

$$V(a_t) = \max_{c_t} u(c_t) + \beta \max_{\lbrace c_v \rbrace_{v=t}^{\infty}} U_{t+1}$$

and the constraints

$$a_{t+1} = a_t - c_t \text{ and } \sum_{v=t}^{\infty} c_v = a_{t+1}.$$

This is the **Bellman's principle of optimality**. We can therefore solve the optimal choice problem at $t$ in two steps: 

1. We determine what is the maximum possible amount of utility that can be attained from $t+1$ onwards, given the remaining amount of resources $a_{t+1}$.

2. We solve for the optimal consumption level at $t$ given $a_t$ as well as $a_{t+1} = a_t - c_t$.

Consequently,

$$V(a_t) = \max_{c_t} u(c_t)+\beta V(a_{t+1}) \text{ s.t. } a_{t+1} = a_t -c_t$$

or without subscripts, using the $+$ sign to indicate variables in the successive period

$$V(a) = \max_{c} u(c)+\beta V(a^+) \text{ s.t. } a^+ = a - c.$$

This is the so-called **Bellman equation** (a.k.a. **functional equation**, since the value of $V$ is the true unknown). The solution $V$ is a fixed point of this equation. Consumption $c$ in this case is a **control variable**, as it controls the reduction of the resource. Finally, $a$ is a **state variable**, as it describes the state of the dynamic problem at time $t$ (the remaining amount of resources in our case).

Since our problem has such a simple structure, an analytical solution exists. Using Lagrange optimization on Bellman equation yields

$$\mathcal{L} = u(c) + \beta V(a^+) + \lambda \Big[a-c-a^+ \Big].$$

The first-order conditions with respect to $c$ and $a^+$ are

$$\frac{\partial \mathcal{L}}{\partial c} = u'(c) - \lambda = 0 \text{ and } \frac{\partial \mathcal{L}}{\partial a^+}= \beta V'(a^+) - \lambda = 0, $$

which results in 

$$ u'(c) = \beta V'(a^+).$$

Since 

$$V(a^+) = \max_{c^+} u(c^+)+\beta V(a^{++}) \text{ s.t. } a^{++} = a^+ - c^+$$

$$ = \max_{a^+} u(a^+ - a^{++})+\beta V(a^{++})$$

we obtain 

$$V'(a^+) = u'(c^+)$$

so that we get 

$$ u'(c) = \beta u'(c^+) \text{ or } c^+ = \beta^{\gamma}c.$$

In order to get a complete solution of the dynamic programm, we make a guess of the functional form of the value function:

$$ V(a) = B\frac{a^{1-\frac{1}{\gamma}}}{1-\frac{1}{\gamma}}.$$

From the first-order condition, we now get

$$c^{-\frac{1}{\gamma}} = \beta B(a^+)^{\frac{1}{\gamma}} \text{ so that } a^+ = (\beta B)^{\gamma}c.$$

Using $c = a-a^+$ we can derive

$$ a^+ = \frac{(\beta B)^{\gamma}a}{1+(\beta B)^{\gamma}} \text{ and } c = \frac{a}{1+(\beta B)^{\gamma}}.$$

The value function must satisfy

$$ B\frac{a^{1-\frac{1}{\gamma}}}{1-\frac{1}{\gamma}} = \max_{c} \frac{c^{1-\frac{1}{\gamma}}}{1-\frac{1}{\gamma}} + \beta B\frac{(a^+)^{1-\frac{1}{\gamma}}}{1-\frac{1}{\gamma}}, $$

so that we obtain

$$ Ba^{1-\frac{1}{\gamma}} = \Bigg[ \frac{a}{1+(\beta B)^{\gamma}} \Bigg]^{1-\frac{1}{\gamma}} + \beta B \Bigg[ \frac{(\beta B)^{\gamma}a}{1+(\beta B)^{\gamma}} \Bigg]^{1-\frac{1}{\gamma}} $$

after substituting the optimal decisions. Dividing by $a^{1-\frac{1}{\gamma}}$ on both sides and rearranging finally leads to $B = \big[1-\beta^{\gamma} \big]^{-\frac{1}{\gamma}}$. Thus we can write

$$V(a) = \Big[1-\beta^{\gamma} \Big]^{-\frac{1}{\gamma}} \frac{a^{1-\frac{1}{\gamma}}}{1-\frac{1}{\gamma}}.$$

Finally, substituting $B$ in the optimal values for savings and consumption we get

$$ a^+ = a^+(a) = a\beta^{\gamma} \text{ and } c=c(a)=a(1-\beta^{\gamma}).$$

The function $c(a)$ is called **policy function** and tells us the value of the control variable for any possible state $a$. In our simple problem, it indicates that the optimal policy is to always eat a constant fraction of the remaining cake. On the other hand, the function $a^+(a)$ describes the development of the state variable.

The code below calculates the dynamics of the cake-eating problem from the policy function $c(a)$, using as parameter values $a_0 = 100$, $\beta = 0.95$, and $\gamma = 0.5$.

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt 

# Model parameters
gamma = 0.5
egam = 1-1/gamma
beta = 0.95
a0 = 100

# Number of periods
TT = 200

# Number of points in the assets grid
NA = 1000

# Calculate the time path of consumption
a_t = np.empty(TT)
c_t = np.empty(TT)
a_t[0] = a0
c_t[0] = a_t[0]*(1-beta**gamma)

for it in range(1, TT):
    a_t[it] = a_t[it-1] - c_t[it-1]
    c_t[it] = a_t[it]*(1-beta**gamma)
    
# Calculate policy and value function
a = np.empty(NA)
c = np.empty(NA)
V = np.empty(NA)

for ia in range(0, NA):
    a[ia] = 1 + ia/NA*a0
    c[ia] = a[ia]*(1-beta**gamma)
    V[ia] = (1-beta**gamma)**(-1/gamma)*a[ia]**egam/egam
```

We can now plot the time path of consumption, together with the policy and the value function as functions of the remaining value of resources $a$:

```python
fig = plt.figure(figsize=(15,15))
fig.suptitle('Analytical Solution to the Cake-Eating Problem')

plt.subplot(321)
plt.plot(range(0, TT), c_t)
plt.xlabel('Time $t$')
plt.ylabel('Consumption $c_t$')
plt.title('Optimal consumption path in the cake-eating problem')

plt.subplot(322)
plt.plot(a, c, color='red')
plt.xlabel('Current level of resources $a$')
plt.ylabel('Policy function $c(a)$')
plt.title('Policy function in the cake-eating problem')

plt.subplot(323)
plt.plot(a, V, color='green')
plt.xlabel('Current level of resources $a$')
plt.ylabel('Value function $V(a)$')
plt.title('Value function in the cake-eating problem')
```
